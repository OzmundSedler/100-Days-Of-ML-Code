# Pearson correlation

```
import pandas as pd
import numpy as np
```

## Foodmart product sales 

```
sales = pd.read_csv('https://raw.githubusercontent.com/OzmundSedler/100-Days-Of-ML-Code/master/week_12/datasets/foodmart.sales.tsv', sep = '\t', header = 0, parse_dates = [2])
```

```
sales.info()
```

```
sales.head()
```

```
sales.describe()
```

```
products = pd.read_csv('https://raw.githubusercontent.com/OzmundSedler/100-Days-Of-ML-Code/master/week_12/datasets/foodmart.products.tsv', sep = '\t', header = 0)
```

```
products.info()
```

```
products.head()
```

```
products.describe()
```

```
sales = sales.merge(products[['product_id', 'product_name']], 
                    on = ['product_id'], how = 'inner')
```

```
sales.head()
```

## Product-product correlation

```
sparse_sales = pd.pivot_table(sales, values='sales', index=['date', 'store_id'],
                     columns=['product_name'], fill_value = 0)
```

```
sparse_sales.head()
```

```
sales_correlation = sparse_sales.corr()
```

```
sales_correlation.head()
```

```
product_name = 'American Chicken Hot Dogs'
sales_correlation[[product_name]].sort_values(product_name, ascending = True).head()
```

```
min_corr = pd.DataFrame(sales_correlation.min())
min_corr.columns = ['min']
min_corr.sort_values(by = 'min').head()
```

```
max_corr = pd.DataFrame(sales_correlation.replace(1,0)).max(axis=1)
max_corr.columns = ['max']
max_corr.sort_values(ascending=False).head()
```

```
product_name = 'Plato French Roast Coffee'
sales_correlation[[product_name]].sort_values(product_name, ascending = False).head()
```
